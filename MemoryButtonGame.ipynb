Below is the hottest game of 2020. A Â£500 reward is given to anyone who is able to win.

A light run to the right will indicate a sequence beggining with a length 1 for you to remember, a second light run to the right shows the sequence has finished and you can now begin reentering the sequence using the corresponding buttons.

If you guess correctly the light run will be to the right and the next level begins with an extra light added onto the sequence, if you are incorrect the light run will be to the left and you must then repeat it with a strike being added onto your sequence.

For complex reasons (laxy programming) the sequence must not be reentered too quickly, nor with you holding the buttons down. Additionally you must complete any guesses up to the length of the actual sequence.

How To play:
- Run the below
- Observe the print statements to see your current score and strikes
- Two light runs to the left indicate that you lose

```python
from time import sleep
from pynq.overlays.base import BaseOverlay
import random

base = BaseOverlay("base.bit")

##useful functions
def playPattern():
    lightRunRight()
    sleep(delay)
    for led in pattern:
        base.leds[led].on()
        sleep(delay)
        base.leds[led].off()
        sleep(delay/2)        
    
def correctGuess():
    sleep(delay)
    lightRunRight()
    guessPattern = []
    

    while len(guessPattern) < len(pattern):
        if (base.buttons[0].read()==1):
            guessPattern.append(0)
            sleep(delay) ## prevent double click bodge but require wait to enter  
        elif (base.buttons[1].read()==1):
            guessPattern.append(1)
            sleep(delay) ## prevent double click bodge but require wait to enter  
        elif (base.buttons[2].read()==1):
            guessPattern.append(2)
            sleep(delay) ## prevent double click bodge but require wait to enter  
        elif (base.buttons[3].read()==1):
            guessPattern.append(3)
            sleep(delay) ## prevent double click bodge but require wait to enter                 
            
    if guessPattern == pattern:
        lightRunRight()
        return True
    else:
        lightRunLeft()
        return False

 
def randomLED():
    return random.randint(0,3)


def lightRunRight():
    for led in reversed(base.leds):
        led.on()
        sleep(lightRunDelay)  
        led.off()
        
def lightRunLeft():
    for led in base.leds:
        led.on()
        sleep(lightRunDelay)  
        led.off()        
        
```

```python
#initialise game
pattern = [randomLED()]
strikes = 0

#initialise play parameters
delay = 0.5
lightRunDelay = 0.2
#color = (color+1) % 8 ##this chnages the colour of the multi color LEDs

#turn LED's all off from previous state
for led in base.leds:
    led.off()  

while(strikes < 3 ):
    print()
    print ('Level: ' + str(len(pattern)))
    playPattern()
    if (correctGuess() == False):
        strikes += 1
        print('Strikes: ' + str(strikes))
    else:
        print('You passed the level.')
        pattern.append(randomLED())

    sleep(delay*2)
    
lightRunLeft()   
lightRunLeft() 
print('You lose.')
```
